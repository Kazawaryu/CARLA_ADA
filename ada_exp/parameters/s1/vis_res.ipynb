```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = [40, 50, 60, 70]

ya90 = [None, 24.0, 20.06, 22.9]
ya80 = [None, 30, 32, 35.7]

yb90 = [6.9, 13.6, 11.9, 16.3]
yb80 = [27.8, 28.9, 28.3, 41.4]

yd90 = [14.9, 15.3, 10.7, 10.5]
yd80 = [20.5, 26.8, 22.2, 21.4]

# Create a figure with two subplots arranged horizontally
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot for 90 sub-items
axs[0].plot(x, ya90, marker='o', mec='r', mfc='w', label='A90', color='blue')
axs[0].plot(x, yb90, marker='o', mec='r', mfc='w', label='B90', color='red')
axs[0].plot(x, yd90, marker='o', mec='r', mfc='w', label='D90', color='green')
axs[0].legend()
axs[0].set_xticks(x)
axs[0].set_xlabel("Discripter Radius")
axs[0].set_ylabel("Acc(%)")
axs[0].set_title("90 Sub-items")

# Plot for 80 sub-items
axs[1].plot(x, ya80, marker='*', ms=10, label='A80', color='blue')
axs[1].plot(x, yb80, marker='*', ms=10, label='B80', color='red')
axs[1].plot(x, yd80, marker='*', ms=10, label='D80', color='green')
axs[1].legend()
axs[1].set_xticks(x)
axs[1].set_xlabel("Discripter Radius")
axs[1].set_ylabel("Acc(%)")
axs[1].set_title("80 Sub-items")

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.3)

# Display the combined plot
plt.show()
```
