```python
import numpy as np
import matplotlib.pyplot as plt

# draw 3d point randomly, and draw the 3d point in 2d
def draw_3d_point(fig, ax):
    ax.set_xlim(-2, 2)
    ax.set_ylim(-2, 2)
    ax.set_zlim(-2, 2)
    # ax.set_xlabel('X')
    # ax.set_ylabel('Y')
    # ax.set_zlabel('Z')

    # draw 3d point randomly
    for i in range(200):
        x = np.random.uniform(-2, 2)
        y = np.random.uniform(-2, 2)
        z = np.random.uniform(-2, 2)
        if x**2 + y**2 + z**2 < 1:
            ax.scatter(x, y, z, color='b')
        else:
            ax.scatter(x, y, z, color='r')

# draw a 3d sphere at (0,0,0)
def draw_3d_sphere(fig, ax):
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x = np.outer(np.cos(u), np.sin(v))
    y = np.outer(np.sin(u), np.sin(v))
    z = np.outer(np.ones(np.size(u)), np.cos(v))
    # 透明展示球面
    ax.plot_surface(x, y, z, color='y', alpha=0.3)

fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')
draw_3d_point(fig, ax)
draw_3d_sphere(fig, ax)

plt.show()
```
